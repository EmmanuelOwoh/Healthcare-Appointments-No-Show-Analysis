```python
# Renaming the columns to snake_case
df.columns = df.columns.str.lower().str.replace('-', '_')
df.rename(columns={
    'patientid': 'patient_id',
    'appointmentid': 'appointment_id',
    'scheduledday': 'scheduled_date',
    'appointmentday': 'appointment_date',
    'neighbourhood': 'clinic',
    'no_show': 'no_show'
}, inplace=True)
```

```python
df.head()
```

```python
# Convert date columns
df['scheduled_date'] = pd.to_datetime(df['scheduled_date']).dt.date
df['appointment_date'] = pd.to_datetime(df['appointment_date']).dt.date
```

```python
df['age'].describe()
```

```python
# Filter invalid ages (0-100 range)
df = df[(df['age'] >= 0) & (df['age'] <= 100)]
```

```python
# Handle missing clinic names
df['clinic'] = df['clinic'].fillna('Unknown')
```

```python
# Calculate wait_days
df['wait_days'] = (pd.to_datetime(df['appointment_date']) - 
                   pd.to_datetime(df['scheduled_date'])).dt.days
```

```python
# Convert no_show to binary (1 = Yes/No-show, 0 = No/Showed up)
df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})
```

```python
#to check for the no_show rate
df['no_show'].value_counts(normalize=True) * 100
```

```python
#to check the number of patients that received an sms
df['sms_received'].value_counts()

#this shows the no-show rate for patients who got sms & those who did not
df.groupby('sms_received')['no_show'].mean() * 100
```

```python
#to show the wait_days by no_show status
df.groupby('no_show')['wait_days'].mean() 
```

```python
# Save as cleaned data
df.to_csv('appointments_cleaned.csv', index=False)
print(f"Cleaned dataset: {len(df)} rows")
```
